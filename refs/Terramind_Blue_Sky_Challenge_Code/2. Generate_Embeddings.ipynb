```python
from torchgeo.datasets import RasterDataset
from torchgeo.samplers import GridGeoSampler
import os
import hashlib
from terratorch.registry import BACKBONE_REGISTRY, FULL_MODEL_REGISTRY
import torch
from shapely.geometry import box
import rasterio
from rasterio.windows import Window
from rasterio.windows import bounds as window_bounds
from google.colab import drive
import geopandas as gpd
from torch.utils.data import DataLoader, BatchSampler
import pandas as pd
drive.mount('/content/drive')
```

```python
# Define Globals
terramind_project = '/content/drive/MyDrive/Terramind_Blue_Sky_Challenge/'
neighborhood_folders = [entry.name for entry in os.scandir(terramind_project) if entry.is_dir()]
neighborhood_folders.remove('archive')
print(neighborhood_folders)
```

```python
class YearRasterDataset(RasterDataset):
    filename_regex = r".*(?P<date>\d{4}).*\.tif"
    date_format = "%Y"

def bbox_id(bounds, ndigits=3):
    b = tuple(round(float(v), ndigits) for v in bounds)  # control float noise
    s = ",".join(map(str, b)).encode("utf-8")
    return hashlib.sha1(s).hexdigest()[:12]


def aligned_safe_roi(tif_path: str, tile_size: int = 224):
  tif_path = neighborhood_path + "/" + os.listdir(neighborhood_path)[0]
  try:
    with rasterio.open(tif_path) as src:
        H, W = src.height, src.width  # rows, cols

        # Largest multiples of tile_size that fit inside the raster
        new_W = (W // tile_size) * tile_size
        new_H = (H // tile_size) * tile_size

        # Center-crop (you can also anchor to top-left; see below)
        col_off = (W - new_W) // 2
        row_off = (H - new_H) // 2

        win = Window(col_off=col_off, row_off=row_off, width=new_W, height=new_H)

        # Convert that pixel window to map bounds (in raster CRS)
        left, bottom, right, top = window_bounds(win, transform=src.transform)

        roi = box(left, bottom, right, top)

        return roi
  except:
      print("Raster too small to make make aligned safe roi")
      return 0

def collate_fn_one_batch(b):
    x = b[0]
    return {
        "image": x["image"],
        "bounds": x["bounds"],
        "crs":x["crs"]
    }

def create_metadata_gdf(grid_sampler, dataset):
  chip_geometries = []
  years = []

  for chip in grid_sampler:
    minx, miny, maxx, maxy = chip[0].start, chip[1].start, chip[0].stop, chip[1].stop
    chip_geometries.append(box(minx, miny, maxx, maxy))
    years.append(chip[2].stop.year)

  gdf = gpd.GeoDataFrame(
      {"chip_id": range(len(chip_geometries)), "year":years},
      geometry=chip_geometries,
      crs="EPSG:32615", # H-Townnnn
  )

  gdf["spatial_id"] = gdf.geometry.bounds.apply(
      lambda row: bbox_id((row.minx, row.miny, row.maxx, row.maxy)),
      axis=1
  )

  gdf["spatial_id_int"], uniques = pd.factorize(gdf["spatial_id"])

  for j in range(10):
    gdf[f'class_{j}'] = 0.0

  for i, chip in enumerate(grid_sampler):
    image = dataset[chip]["image"]
    total_pixels = image.numel()
    unique_classes, counts = torch.unique(image, return_counts=True)

    for class_val, count in zip(unique_classes, counts):
      percentage = (count.item() / total_pixels) * 100
      # Assign percentage to the corresponding column in metadata
      gdf.loc[i, f'class_{int(class_val.item())}'] = percentage

  print(f"Made {len(chip_geometries)} chips")

  return gdf


def tile_rasters(raster_folder):
  roi = aligned_safe_roi(raster_folder)
  dataset = YearRasterDataset(raster_folder)
  try:
    sampler = GridGeoSampler(dataset, size=224, stride=224, roi=roi)
    metadata = create_metadata_gdf(sampler, dataset)
    loader = DataLoader(
      dataset,
      sampler=sampler,
      batch_size=1,
      collate_fn=collate_fn_one_batch,
      num_workers=0,
      )
    print(f"Successfully loaded {len(dataset)} rasters")
    return metadata, loader
  except:
    return [], []


def get_embeddings(loader, model):

  all_embeddings = []

  with torch.no_grad():
    for batch in loader:
      LULC_input = batch['image'].to(device)
      out = model(LULC=LULC_input)
      embeddings = out[-1]
      all_embeddings.append(embeddings.cpu())

  print(f"Successfully extracted embeddings from Terramind")

  return all_embeddings

def save_embeddings(all_embeddings, metadata, neighborhood_path):
  X = torch.cat(all_embeddings, dim=0)
  T = len(metadata['year'].unique())
  N = X.shape[0]
  tile = N // T
  emb = X.view(T, tile, X.shape[1], X.shape[2])  # (time, tile, patch, embed)
  torch.save(
      {
          "embeddings": emb
      },
      neighborhood_path + "/terramind_lulc_embeddings.pt"
  )
  metadata.to_file(neighborhood_path + "/lulc_classes_embeddings_metadata.geojson", driver="GeoJSON")
  print(f"Successfully saved embeddings and metadata to {neighborhood_path}")
  print(f"Embedding shape {emb.shape}")
  return
```

```python
if torch.cuda.is_available():
    device = 'cuda'
elif torch.backends.mps.is_available():
    device = 'mps'
else:
    device = 'cpu'

all_embeddings = []

model = BACKBONE_REGISTRY.build(
      'terramind_v1_base',
      pretrained=True,
      modalities=["LULC"])
model.eval()
model.to(device);
```

```python
for neighborhood in neighborhood_folders:
  print(f"Processing {neighborhood}")
  neighborhood_path = os.path.join(terramind_project, neighborhood)
  metadata, loader = tile_rasters(neighborhood_path)
  if len(metadata) == 0:
    print(f" Rasters too small in {neighborhood_path}, skipping...")
    continue
  embeddings = get_embeddings(loader, model)
  save_embeddings(embeddings, metadata, neighborhood_path)
```
