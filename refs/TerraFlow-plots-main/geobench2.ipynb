# Generate Table

```python
datasets = ["kuro_siwo", "pastis", "biomassters", "dynamic_earthnet"]
models = ["tm_tiny", "tf_tiny_rs", "tf_tiny_tds", "tm_base", "tf_base_rs", "tf_base_tds"]

all_data = {
    "tm_tiny": data_tm_tiny,
    "tf_tiny_rs": data_tf_tiny_rs,
    "tf_tiny_tds": data_tf_tiny_tds,
    "tm_base": data_tm_base,
    "tf_base_rs": data_tf_base_rs,
    "tf_base_tds": data_tf_base_tds,
}

# -------------------------------------------------
# Plain mean / std
# -------------------------------------------------
mean_per_dataset = {
    m: {d: np.mean(data[d]) for d in datasets} for m, data in all_data.items()
}
std_per_dataset = {
    m: {d: np.std(data[d]) for d in datasets} for m, data in all_data.items()
}

# -------------------------------------------------
# Build DataFrames (×100)
# -------------------------------------------------
mean_df = pd.DataFrame(
    {m: [mean_per_dataset[m][d] * 100 for d in datasets] for m in models},
    index=datasets,
).T

std_df = pd.DataFrame(
    {m: [std_per_dataset[m][d] * 100 for d in datasets] for m in models},
    index=datasets,
).T

# -------------------------------------------------
# Combine into "mean ± std"
# -------------------------------------------------
mean_std_df = mean_df.copy()
for m in models:
    for d in datasets:
        mean_std_df.loc[m, d] = f"{mean_df.loc[m, d]:.1f} ± {std_df.loc[m, d]:.1f}"

# -------------------------------------------------
# Rename rows and columns
# -------------------------------------------------
row_rename = {
    "tm_tiny": "TerraMind-V1-Tiny",
    "tf_tiny_rs": "TerraFlow-Tiny-RS",
    "tf_tiny_tds": "TerraFlow-Tiny-TDS",
    "tm_base": "TerraMind-V1-Base",
    "tf_base_rs": "TerraFlow-Base-RS",
    "tf_base_tds": "TerraFlow-Base-TDS",
}

col_rename = {
    "kuro_siwo": "KuroSiwo",
    "pastis": "PASTIS",
    "dynamic_earthnet": "DynamicEarthNet",
    "biomassters": "BioMassters",
}

mean_df.rename(index=row_rename, columns=col_rename, inplace=True)
std_df.rename(index=row_rename, columns=col_rename, inplace=True)
mean_std_df.rename(index=row_rename, columns=col_rename, inplace=True)

# -------------------------------------------------
# Styling
# -------------------------------------------------
def highlight_best_and_second(col):
    means = mean_df[col.name]
    order = means.sort_values(ascending=(col.name == "BioMassters"))
    best, second = order.iloc[0], order.iloc[1]
    return [
        "font-weight: bold;" if means.loc[i] == best
        else "text-decoration: underline;" if means.loc[i] == second
        else ""
        for i in col.index
    ]

styled = mean_std_df.style.apply(highlight_best_and_second, axis=0)
display(styled)
```

```python
# ================================
# HARD-CODE MODEL PARAMETERS
# ================================
model_params = {
    "Tiny": "8M",
    "Base": "100M",
}

mean_df = mean_df.copy()
std_df = std_df.copy()

# Hard-code Params column EXACTLY by index
mean_df.insert(
    0,
    "Params",
    [model_params.get(idx, "") for idx in mean_df.index]
)
std_df.insert(0, "Params", [""] * len(std_df))


# ================================
# LATEX HIGHLIGHT FUNCTION
# ================================
def latex_highlight(mean_df, std_df):
    out = mean_df.copy()

    for col in mean_df.columns:
        if col == "Params":
            continue

        # numeric ranking (CRITICAL)
        vals = mean_df[col].astype(float)

        ascending = (col == "BioMassters")
        order = vals.sort_values(ascending=ascending)

        best = order.index[0]
        second = order.index[1]

        for row in mean_df.index:
            val = f"${mean_df.loc[row, col]:.1f} \\pm {std_df.loc[row, col]:.1f}$"

            if row == best:
                val = f"\\boldmath{{{val}}}"
            elif row == second:
                val = f"\\underline{{{val}}}"

            out.loc[row, col] = val

    return out.astype(str)


# ================================
# GENERATE LATEX
# ================================
latex_df = latex_highlight(mean_df, std_df)

latex_code = latex_df.to_latex(
    escape=False,
    column_format="l" + "c" * mean_df.shape[1],
    caption="Mean $\\pm$ Std results across datasets",
    label="tab:results"
)

print(latex_code)
```
